#The Pearson correlation matrix
- Is a table that displays the correlation coefficients between pairs of variables. The correlation coefficient measures the strength and direction of a linear relationship between two variables, ranging from -1 to 1. A value of -1 indicates a perfect negative correlation, 0 indicates no correlation, and 1 indicates a perfect positive correlation. The Pearson correlation matrix is often used in data analysis to identify patterns and relationships between variables.

```python
df.corr(method ='kendall')
```

#The Kendall correlation matrix
- Is a table that displays the Kendall rank correlation coefficients between pairs of variables. The Kendall correlation coefficient measures the strength and direction of a monotonic relationship between two variables, which is a relationship that increases or decreases in a consistent manner but not necessarily at a constant rate. 

- The coefficient ranges from -1 to 1, where -1 indicates a perfect negative monotonic correlation, 0 indicates no monotonic correlation, and 1 indicates a perfect positive monotonic correlation. The Kendall correlation matrix is often used when the relationship between variables is non-linear or the data is ranked or ordinal





```python
df.corr(method ='spearman')
```

#The Spearman correlation matrix 
- Is a table that displays the Spearman rank correlation coefficients between pairs of variables. The Spearman correlation coefficient measures the strength and direction of a monotonic relationship between two variables, which is a relationship that increases or decreases in a consistent manner but not necessarily at a constant rate. 

- The coefficient ranges from -1 to 1, where -1 indicates a perfect negative monotonic correlation, 0 indicates no monotonic correlation, and 1 indicates a perfect positive monotonic correlation. The Spearman correlation matrix is often used when the relationship between variables is non-linear or the data is ranked or ordinal, similar to the Kendall correlation matrix.

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```
